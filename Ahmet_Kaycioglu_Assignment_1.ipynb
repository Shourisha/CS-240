```python
fmarno_data = resp.fmarno.value_counts()
fmarno_data
```

```python
fmarno_hist = thinkstats2.Hist(resp.fmarno, label='# of marriages of respondents')
thinkplot.hist(fmarno_hist)
thinkplot.config(xlabel='# of marriages', ylabel='Respondents')
```

```python
neverMarried = resp[resp.fmarno==0]
married      = resp[resp.fmarno!=0]
married.totincr.value_counts()
```

```python
neverMarried_hist = thinkstats2.Hist(neverMarried.totincr, label='Never Married')
married_hist = thinkstats2.Hist(married.totincr, label='Married')
```

```python
width=0.4
thinkplot.PrePlot(2)
thinkplot.hist(neverMarried_hist, align='right', width=width)
thinkplot.hist(married_hist, align='left', width = width)
thinkplot.Config(xlabel='Income', ylabel='# of families')
```

```python
print ('Stats for Never Married:\n Minimum:' + str(neverMarried.totincr.min())+ '\n Maximum:' + str(neverMarried.totincr.max()) + '\n Mean:' +
      str(neverMarried.totincr.mean()) + '\n Variance:' + str(neverMarried.totincr.var()) + '\n Standard Deviation:' + str(neverMarried.totincr.std()))
```

```python
print ('Stats for Married:\n Minimum:' + str(married.totincr.min())+ '\n Maximum:' + str(married.totincr.max()) + '\n Mean:' +
      str(married.totincr.mean()) + '\n Variance:' + str(married.totincr.var()) + '\n Standard Deviation:' + str(married.totincr.std()))
```
